# Loss Plot for Contrastive Pre_trained

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_accuracy(results1, results2, y_label, multiplier = 1.96):
    f, a = plt.subplots()
    plt.plot(np.arange(0,results1.shape[1]), np.mean(results1,0), label='Train')
    plt.fill_between(np.arange(0,results1.shape[1]),
                     np.mean(results1,0)+(multiplier*np.std(results1,0)/np.sqrt(results1.shape[0])), 
                     np.mean(results1,0)-(multiplier*np.std(results1,0)/np.sqrt(results1.shape[0])), alpha=0.5)
    plt.plot(np.arange(0,results2.shape[1]), np.mean(results2,0), label='Validation')
    plt.fill_between(np.arange(0,results2.shape[1]), 
                     np.mean(results2,0)-(multiplier*np.std(results2,0)/np.sqrt(results2.shape[0])),
                     np.mean(results2,0)+(multiplier*np.std(results2,0)/np.sqrt(results2.shape[0])), alpha=0.5)
    # plt.ylim([0.0,1.0])
    plt.ylabel(y_label)
    plt.xlabel('Epoch')
    plt.legend(loc='best')
    plt.grid()
    plt.title('Contrastive Learn Loss Plot')
    # plt.legend(['Adam','RMSprop','SGD'])
    # plt.xticks(ticks=np.arange(0,results1.shape[1]), labels=np.arange(0,results1.shape[0]+1,10))
    plt.show()
```

```python
FDtrain_loss = np.loadtxt("./ContrastiveLearnedModel_MonteCarloResults/train_loss.txt")
FDvalidation_loss = np.loadtxt("./ContrastiveLearnedModel_MonteCarloResults/validation_loss.txt")
plot_accuracy(FDtrain_loss, FDvalidation_loss, y_label = 'Loss')
```

```python
FDtrain_loss = np.loadtxt("./CLM_CorrectedValResults/train_loss.txt")
FDvalidation_loss = np.loadtxt("./CLM_CorrectedValResults/validation_loss.txt")
plot_accuracy(FDtrain_loss, FDvalidation_loss, y_label = 'Loss')
```

```python
train_loss = np.loadtxt("./CLM_CorrectedValResults/NewRun/train_loss.txt")
validation_loss = np.loadtxt("./CLM_CorrectedValResults/NewRun/validation_loss.txt")
epochs = range(1, 200+1)

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 2)
plt.plot(epochs, train_loss, label='Train Loss')
plt.plot(epochs, validation_loss, label='Val Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Contrastive Learn Loss Plot batch size 16')

plt.legend()
plt.grid()
plt.show()
```
